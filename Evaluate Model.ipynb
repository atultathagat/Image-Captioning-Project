```python
with open('Flickr8k_text/Flickr_8k.testImages.txt','r') as file_obj:
    data_test=file_obj.readlines()
```

```python
references=[]
candidates=[]
for cnt,img in enumerate(data_test):
    img=img.split('.')[0]
    caption_actuals=img_captions_dict[img]
    caption_predicted=predict(img)
    print('Caption for image-',cnt+1,'predicted')
    reference=[]
    for caption in caption_actuals:
        reference.append(caption.split())
    references.append(reference)
    candidates.append(caption_predicted.split())
```

```python
from nltk.translate.bleu_score import corpus_bleu
```
