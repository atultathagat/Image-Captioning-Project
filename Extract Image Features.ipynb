```python
import os
from tensorflow.keras.preprocessing.image import load_img,img_to_array
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.models import Model
import pickle
```

```python
src='Flickr8k_Dataset\Flicker8k_Dataset'
```

```python
vgg=VGG16()
```

```python
model=Model(inputs=vgg.input,outputs=vgg.layers[-2].output)
```

```python
model.summary()
```

```python
dict_img_features={}
for cnt,img_name in enumerate(os.listdir(src)):
        img_path=os.path.join(src,img_name)
        img=load_img(img_path,target_size=(224,224))
        img_arr=img_to_array(img)
        img_arr=preprocess_input(img_arr)
        img_arr=img_arr.reshape(-1,224,224,3)
        img_features=model.predict(img_arr)[0]
        img_name=img_name.split('.')[0].strip()
        dict_img_features[img_name]=img_features
        print('Features of Image - ',cnt+1,'extracted.')
```

```python
dict_img_features
```

```python
with open('image_features_pkl','wb') as file:
    pickle.dump(dict_img_features,file)
```
